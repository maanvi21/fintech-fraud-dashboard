```python
import pandas as pd

# Load datasets
insurance = pd.read_csv('insurance_data.csv')
vendor = pd.read_csv('vendor_data.csv')
employee = pd.read_csv('employee_data.csv')

# Preview
print("Insurance Data:")
display(insurance.head())

print("Vendor Data:")
display(vendor.head())

print("Employee Data:")
display(employee.head())
```

```python
# Basic info
print("Insurance Info")
print(insurance.info())

print("\nVendor Info")
print(vendor.info())

print("\nEmployee Info")
print(employee.info())
```

```python
print("Missing Values (Insurance):")
print(insurance.isnull().sum())

print("\nMissing Values (Vendor):")
print(vendor.isnull().sum())

print("\nMissing Values (Employee):")
print(employee.isnull().sum())
```

```python
insurance.describe()
```

```python
# Drop duplicates
insurance = insurance.drop_duplicates()
vendor = vendor.drop_duplicates()
employee = employee.drop_duplicates()

# Fill missing values (simple approach)
insurance = insurance.fillna(0)
vendor = vendor.fillna(0)
employee = employee.fillna(0)
```

```python
# Start with insurance as base
df = insurance.copy()

# Add vendor details using vendor_id
df = df.merge(vendor, on='VENDOR_ID', how='left')

# Add employee details using agent_id
df = df.merge(employee, on='AGENT_ID', how='left')

# Remove duplicate columns if created
df = df.loc[:, ~df.columns.duplicated()]

# Preview
df.head()
```

```python
# Save as CSV
df.to_csv('merged_insurance_data.csv', index=False)
from google.colab import files
files.download('merged_insurance_data.csv')
```

```python

def find_date_columns(df, name):
    print(f"\n{name} dataset:")
    date_cols = []

    for col in df.columns:
        try:
            # Try converting sample values to datetime
            pd.to_datetime(df[col].dropna().sample(min(10, len(df))), errors='raise')
            date_cols.append(col)
        except:
            pass

    print("Detected date columns:", date_cols)

find_date_columns(insurance, "Insurance")
find_date_columns(vendor, "Vendor")
find_date_columns(employee, "Employee")
```

```python
# Insurance dates
for col in ['TXN_DATE_TIME', 'POLICY_EFF_DT', 'LOSS_DT', 'REPORT_DT']:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Employee date
df['DATE_OF_JOINING'] = pd.to_datetime(df['DATE_OF_JOINING'], errors='coerce')
```

```python
df['policy_age_days'] = (df['LOSS_DT'] - df['POLICY_EFF_DT']).dt.days
```

```python
df['report_delay_days'] = (df['REPORT_DT'] - df['LOSS_DT']).dt.days
```

```python
df['processing_delay'] = (df['TXN_DATE_TIME'] - df['REPORT_DT']).dt.days
```

```python
df['agent_experience_days'] = (df['LOSS_DT'] - df['DATE_OF_JOINING']).dt.days
```

```python
# normalization of numerical columns
num_cols = [
    'CLAIM_AMOUNT',
    'PREMIUM_AMOUNT',
    'policy_age_days',
    'report_delay_days',
    'processing_delay',
    'agent_experience_days'
]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
# visualizations
import matplotlib.pyplot as plt

plt.hist(df['CLAIM_AMOUNT'], bins=50)
plt.title("Claim Amount Distribution")
plt.xlabel("Claim Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df['policy_age_days'], df['CLAIM_AMOUNT'])
plt.xlabel("Policy Age (days)")
plt.ylabel("Claim Amount")
plt.title("Policy Age vs Claim Amount")
plt.show()
```

```python
plt.hist(df['report_delay_days'], bins=30)
plt.title("Report Delay Distribution")
plt.show()
```

```python
# Clean column names
df.columns = df.columns.str.strip().str.lower()

# Convert to string (VERY IMPORTANT)
df['vendor_id'] = df['vendor_id'].astype(str)

# Remove fake '0' values
df = df[df['vendor_id'] != '0']

# Handle missing properly
df['vendor_id'] = df['vendor_id'].replace('nan', 'Unknown')

# Recompute vendor claim count
df['vendor_claim_count'] = df.groupby('vendor_id')['vendor_id'].transform('count')

# Check result
print(df['vendor_id'].value_counts().head(10))
```

```python
import matplotlib.pyplot as plt

top_vendors = df['vendor_id'].value_counts().head(10)

top_vendors.plot(kind='bar')
plt.title("Top Vendors by Claim Count (Cleaned)")
plt.xticks(rotation=45)
plt.show()
```
